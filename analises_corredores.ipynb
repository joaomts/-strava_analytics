## Bibliotecas

```python
import pandas as pd
```

## Manipulação dos Dados

### Importando dados do percurso

```python
df_gpx = pd.read_csv(
    "tabela_final_lsmision_35km_ultima.csv", sep=";", index_col="Unnamed: 0")
df_gpx.head()
```

```python
# Converter colunas de números com vírgula para ponto e tipo float
cols_to_convert = ['latitude', 'longitude', 'elevacao', 'diff_elev', 'distancias', 'soma_elev', 'soma_dist', 'desnivel_positivo', 'desnivel_negativo']
for col in cols_to_convert:
    df_gpx[col] = df_gpx[col].str.replace(',', '.', regex=False).astype(float)
df_gpx.head()
```

```python
df_grouped = df_gpx.groupby('km_group').agg({
    'elevacao': 'first',
    'distancias': 'sum',
    'soma_elev': 'last',
    'soma_dist': 'last',
    'desnivel_positivo': 'sum',
    'desnivel_negativo': 'sum'
}).reset_index()
df_grouped.head()
```

### Importando dados dos Atletas

```python
df_atletas_2023 = pd.read_csv("final_completo_dados_2023_FINAL_SEM_ERRO.csv", sep=',')
df_atletas_2023.head()
```

```python
df_atletas_2023_selecionado = df_atletas_2023[['Nome Atleta', 'Volta', 'Tempo', 'FC', 'sexo', 'faixa_etaria', 'peso']]
df_atletas_2023_selecionado.head()
```

```python
df_atletas_2022 = pd.read_csv("final_completo_dados_2022_P1.csv", sep=',')
df_atletas_2022.head()
```

```python
df_atletas_2022_selecionado = df_atletas_2022[['Nome Atleta', 'Volta', 'Tempo', 'FC', 'sexo', 'faixa_etaria', 'peso']]
df_atletas_2022_selecionado.head()
```

```python
df_atletas_concat = pd.concat([df_atletas_2023_selecionado, df_atletas_2022_selecionado], ignore_index=True)
df_atletas_concat
```
