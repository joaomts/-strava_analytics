```python
import pandas as pd
```

```python
df_gpx = pd.read_csv(
    "tabela_final_lsmision_35km_ultima.csv", sep=";", index_col="Unnamed: 0"
)
df_gpx.head()
```

```python
df_grouped = df_gpx.groupby('km_group').agg({
    'elevacao': 'first',
    'distancias': 'sum',
    'soma_elev': 'last',
    'soma_dist': 'last',
    'desnivel_positivo': 'sum',
    'desnivel_negativo': 'sum'
}).reset_index()
df_grouped.head()
```
